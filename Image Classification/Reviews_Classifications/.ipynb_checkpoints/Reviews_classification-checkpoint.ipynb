```python
#Remove null values
data = data.dropna()
```

```python
# get list of stop words ()
stopwords = list(STOP_WORDS)
# get list of punctuations and add new line(\n) to the list 
punctuation = punctuation + '\n'

def remove_stopwords_punctuation(review):
    text = ' '.join([word for word in review.split() if word not in stopwords and word not in punctuation])
    return text
```

```python
# function to find polarity of text (polarity - how postive the review/how negative the review)
def find_polarity(review):
    text = remove_stopwords_punctuation(review.lower())
    return TextBlob(text).sentiment.polarity
```

```python
data['Sentiment_Polarity'] = data['Review'].apply(find_polarity)
data
```

```python
sns.distplot(data['Sentiment_Polarity'])
```

```python
sns.barplot(x='Star Ratings', y='Sentiment_Polarity', data=data)
```

```python
negative = data[data.Sentiment_Polarity < 0]
negative
```

```python
positive = data[data.Sentiment_Polarity > 0]
positive
```

```python
Status = []

# classify if the review is positive or negative. positive = 1, negative = 0
for index, row in data.iterrows():
    if row.Sentiment_Polarity >= 0:
        Status.append(1)
    else:
        Status.append(0)
```

```python
new_data = data
new_data['Status'] = Status
```

```python
item_code = []
ratings = []
positive = []
negative = []
category = []

gb = new_data.groupby('Item code')    
for x in gb.groups:
    grp = gb.get_group(x)
    rating = grp['Star Ratings'].sum() / len(grp)
    positive_count = 0
    negative_count = 0
    st = list(grp['Status'])
    ct = list(grp['Product Name'])
    for i in st:
        if(i == 1):
            positive_count += 1
        else:
            negative_count += 1
   
    item_code.append(x)
    ratings.append(rating)
    positive.append(positive_count)
    negative.append(negative_count)
    category.append(ct[0])
```

```python
#create new df 
df = pd.DataFrame({'Product Category':category})
df['Item Code'] = item_code
df['Ratings'] = ratings
df['Positive Count'] = positive
df['Negative Count'] = negative
```

```python
# Save dataset with review status
df.to_csv("dataset_with_status.csv", sep=',', encoding='utf-8')
```
